```python
%%capture --no-stdout
patient=np.zeros((570000,2),dtype=np.int32)
for i in range(1,10001):
    if i%500==0:
        print(i)
    index=57*(i-1)
    df=df_patient[df_patient['Physician_ID']==('axt'+str(i).zfill(5))].set_index('Date_of_Visit')
    l_week=list(df[df['Gender']=='M'].index.week)
    l_year=list(df[df['Gender']=='M'].index.year)
    for j in range(len(l_week)):
        if l_year[j]==2019:
            patient[index+l_week[j]-1][0]+=1
        else:
            patient[index+l_week[j]+51][0]+=1
    l_week=list(df[df['Gender']=='F'].index.week)
    l_year=list(df[df['Gender']=='F'].index.year)
    for j in range(len(l_week)):
        if l_year[j]==2019:
            patient[index+l_week[j]-1][1]+=1
        else:
            patient[index+l_week[j]+51][1]+=1
```

```python
def trim(x):
    if x<=0.4:
        return 0
    elif x<round(x)+0.2 and x>round(x)-0.2:
        return x
    else :
        return round(x)
```

```python
%%capture --no-stdout
l1=np.zeros((500))
feature=np.zeros((500,3))
for i in range(100):
    lg=LinearRegression()
    xgb=xgboost.XGBRegressor(max_depth=4,max_leaves=2,gamma=0.05)
    df_temp=df[df['Physician_ID']=='axt'+str(random.randint(1,10000)).zfill(5)]
    df_temp['Market_Rx_lag1']=df_temp['Market_Rx'].shift(1).fillna(0.0)
    df_temp['Market_Rx_lag2']=df_temp['Market_Rx'].shift(2).fillna(0.0)
    df_temp['Market_Rx_lag3']=df_temp['Market_Rx'].shift(3).fillna(0.0)
    df_temp['Market_Rx_lag4']=df_temp['Market_Rx'].shift(4).fillna(0.0)
    df_temp['Market_Rx_lag5']=df_temp['Market_Rx'].shift(5).fillna(0.0)
    df_temp['Brand_Rx_lag1']=df_temp['Brand_Rx'].shift(1).fillna(0.0)
    df_temp['Brand_Rx_lag2']=df_temp['Brand_Rx'].shift(2).fillna(0.0)
    df_temp['Brand_Rx_lag3']=df_temp['Brand_Rx'].shift(3).fillna(0.0)
    df_temp['Brand_Rx_lag4']=df_temp['Brand_Rx'].shift(4).fillna(0.0)
    df_temp['Brand_Rx_lag5']=df_temp['Brand_Rx'].shift(5).fillna(0.0)
    train=df_temp[df_temp.index<='2019-12-27']
    test=df_temp[df_temp.index>'2019-12-27']
    lg.fit(train.iloc[:,2:],train['Brand_Rx'])
    y_hat1=lg.predict(df_temp.iloc[:,2:])
    xgb.fit(train.iloc[:,2:],train['Brand_Rx']-y_hat1[:-5])
    y_hat2=xgb.predict(test.iloc[:,2:])
    test['y_pred']=y_hat1[-5:]+y_hat2
    test['y_pred']=test['y_pred'].apply(trim)
    l1[5*i:5+5*i]=np.array(test['Brand_Rx']-test['y_pred'])
```

```python
%%capture --no-stdout
feature=np.zeros((10000,3))
for i in range(1,10001):
    if (i)%500==0:
        print(i)
    # lg=Ridge()
    model=RandomForestRegressor(min_samples_split=3,min_samples_leaf=2,max_depth=4,n_jobs=-1)
    df_temp=df[df['Physician_ID']=='axt'+str(i).zfill(5)]
    df_temp['Market_Rx_lag1']=df_temp['Market_Rx'].shift(1).fillna(0.0)
    df_temp['Market_Rx_lag2']=df_temp['Market_Rx'].shift(2).fillna(0.0)
    df_temp['Market_Rx_lag3']=df_temp['Market_Rx'].shift(3).fillna(0.0)
    df_temp['Market_Rx_lag4']=df_temp['Market_Rx'].shift(4).fillna(0.0)
    df_temp['Market_Rx_lag5']=df_temp['Market_Rx'].shift(5).fillna(0.0)
    df_temp['Brand_Rx_lag1']=df_temp['Brand_Rx'].shift(1).fillna(0.0)
    df_temp['Brand_Rx_lag2']=df_temp['Brand_Rx'].shift(2).fillna(0.0)
    df_temp['Brand_Rx_lag3']=df_temp['Brand_Rx'].shift(3).fillna(0.0)
    df_temp['Brand_Rx_lag4']=df_temp['Brand_Rx'].shift(4).fillna(0.0)
    df_temp['Brand_Rx_lag5']=df_temp['Brand_Rx'].shift(5).fillna(0.0)
    model.fit(df_temp.iloc[:,2:],df_temp['Brand_Rx'])
    feature[i-1]=model.feature_importances_[1:4]
```

```python
pred=pd.DataFrame(feature)
pred[0]=pred[0].apply(abs)
pred[1]=pred[1].apply(abs)
pred[2]=pred[2].apply(abs)
pred['y_pred']=np.argmax(feature,axis=1)
```

```python
temp=pd.DataFrame()
temp['Sales_Rep_Calls']=call
temp['catagory']=df_hcp.groupby('Physician_ID').first().Physician_Segment.values
```

```python
def valid_call(a,b):
    if b=='3-Low' and a>=12:
        return 0
    elif b=='2-Medium' and a>=24:
        return 0
    elif b=='1-High' and a>=48:
        return 0
    else:
        return 1
```

```python
pred['valid_call']=np.array(temp.apply(lambda x:valid_call(x.Sales_Rep_Calls,x.catagory),axis=1))
```

```python
def transfer(a,b):
    if a>b:
        return 1
    else:
        return 2
```

```python
index=pred[np.logical_and(pred['valid_call']==0,pred['y_pred']==0)].index
```

```python
l=pred[np.logical_and(pred['valid_call']==0,pred['y_pred']==0)].apply(lambda x:transfer(x[1],x[2]),axis=1)
```

```python
for i in index:
    pred['y_pred'][i]=l[i]
```
